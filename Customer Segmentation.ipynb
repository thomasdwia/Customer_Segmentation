```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# PREPROCESSING DATA

```python
df = pd.read_csv(r"D:\UGM\Sourcecode\data\Mall_Customers.csv", usecols=['Gender','Age','Annual Income (k$)','Spending Score (1-100)'])
df.head()
```

```python
plt.title("ANNUAL INCOME (k$)")
sns.violinplot(y=df["Annual Income (k$)"])
plt.show()
```

```python
plt.title("SPENDING SCORE (1-100)")
sns.violinplot(y=df["Spending Score (1-100)"])
plt.show()
```

# ELBOW CLUSTER

```python
from sklearn.cluster import KMeans
inertias = []

for i in range(1,11):
    kmeans = KMeans(n_clusters = i)
    kmeans.fit(df.iloc[:,1:])
    inertias.append(kmeans.inertia_)
plt.plot(range(1,11), inertias, marker='o')
plt.title("ELBOW METHODS")
plt.xlabel("Jumlah Cluster")
plt.ylabel("Inertia")
plt.show()
```

# CLUSTERING

```python
kmeans = KMeans(n_clusters = 5)
kmeans.fit(df.iloc[:,1:])
kmeans_prediksi = kmeans.predict(df.iloc[:,1:])

plt.scatter(df["Spending Score (1-100)"],df["Annual Income (k$)"], c = kmeans_prediksi, s=50, cmap='viridis')
plt.title("HASIL CLUSTER")
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1-100)")
plt.show()
```
